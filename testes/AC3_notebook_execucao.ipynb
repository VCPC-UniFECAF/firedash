# AC3 — Execução de testes (Jupyter)

As células abaixo chamam os mesmos scripts da pasta `testes/` para você **copiar a saída** ou exportar o notebook com resultados.

**Requisito:** abrir o Jupyter a partir da raiz do repositório `firedash` ou da pasta `testes/` (o código localiza `dashboard/market_utils.py`).


```python
import os
import subprocess
import sys
from pathlib import Path


def raiz_firedash() -> Path:
    for base in [Path.cwd(), *Path.cwd().parents]:
        if (base / "dashboard" / "market_utils.py").is_file():
            return base
    raise RuntimeError(
        "Não encontrei dashboard/market_utils.py. "
        "Inicie o Jupyter na raiz do clone ou faça os.chdir para lá."
    )


ROOT = raiz_firedash()
os.chdir(ROOT)
print("ROOT =", ROOT)
```

## Smoke das fontes (UOL, Yahoo, yfinance, `ifix.csv`)

```python
subprocess.run(
    [sys.executable, str(ROOT / "testes" / "smoke_fontes_dashboard.py")],
    cwd=str(ROOT),
    check=False,
)
```

## Passo 7 — Checagens HTTP (`scripts/qa_http_checks.py` via wrapper)

```python
subprocess.run(
    [sys.executable, str(ROOT / "testes" / "rodar_http_checks_ac3.py")],
    cwd=str(ROOT),
    check=False,
)
```

## Passo 8 — pytest (`pip install -r requirements-dev.txt` se necessário)

```python
subprocess.run(
    [sys.executable, str(ROOT / "testes" / "rodar_pytest_ac3.py")],
    cwd=str(ROOT),
    check=False,
)
```
